```python
import numpy as np 
import pandas as pd
np.set_printoptions(suppress=True)

data_tot = np.load('raw_data/vsdata.npy')
m,N = data_tot.shape
n = int(N/8)
data_tot = data_tot.reshape(12,8,n)
data_mean = data_tot.mean(axis=2).T
data_std = data_tot.std(axis=2).T

# formatted_data = np.char.add(np.char.add(data_mean.round(2).astype(str), '±'), data_std.round(2).astype(str))
formatted_data =data_mean.round(4).astype(str)
# to DataFrame
column_names = ['nsga2_time','nsga2_hv','nsga3_time','nsga3_hv','moead_time','moead_hv','rvea_time','rvea_hv','NI-bSB_time','NI-bSB_hv','NI-dSB_time','NI-dSB_hv']  
row_names = ['n=10,d=0.5','n=10,d=1.0','n=20,d=0.5','n=20,d=1.0','n=100,d=0.5','n=100,d=1.0','n=200,d=0.5','n=200,d=1.0'] 
df = pd.DataFrame(formatted_data, index = row_names,columns=column_names)
# df.to_csv('table1.csv', index=True,encoding='utf-8-sig')
print(df)
```
